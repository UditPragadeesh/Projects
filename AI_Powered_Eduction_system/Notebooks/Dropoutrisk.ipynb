```python
def outlier_count(df,cols):
    outlier_dict={}
    for col in cols:
        q1 = np.percentile(data[col],25)
        q3 = np.percentile(data[col],75)
        iqr = q3-q1
        lower_fence = q1-(1.5*iqr)
        upper_fence = q3+(1.5*iqr)
        outliers = [x for x in df[col] if (x<lower_fence) or (x>upper_fence)]
        outlier_dict[col]=len(outliers)
    return(outlier_dict)

print(outlier_count(data,num_cols))

```

```python
cols = list(dropout_df.columns)
cols.remove('Dropout')
X=dropout_df[cols].to_numpy()
y = dropout_df['Dropout'].to_numpy()
```

```python
import torch
from torch.utils.data import TensorDataset,DataLoader
from torch import nn
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from torch import optim
from sklearn.preprocessing import StandardScaler
```

```python
smote =SMOTE(random_state=2)
X_res,y_res = smote.fit_resample(X,y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_res,y_res,test_size=0.2)
```

```python
scaler = StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
train_dataset = TensorDataset(torch.tensor(x_train).float(),torch.tensor(y_train).float())
train_loader = DataLoader(train_dataset,batch_size=32,shuffle=True)
```

```python
test_dataset = TensorDataset(torch.tensor(x_test).float(),torch.tensor(y_test).float())
test_loader = DataLoader(test_dataset,batch_size=16,shuffle=True)
```

```python
features,target =next(iter(train_loader))
target = target.view(-1,1)
print(features.shape)
print(target.shape)
```

```python
model = nn.Sequential(nn.Linear(17,32),nn.ReLU(),nn.Dropout(0.3),nn.Linear(32,1),nn.Sigmoid())
criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(),lr=0.001,weight_decay=0.0005)
train_losses = []
test_losses = []
```

```python
best_val_loss = float('inf')
patience = 30
trigger_times = 0
```

```python
losses ={}
for epoch in range(1000):
    model.train(True)
    training_loss = 0.0
    for features,target in train_loader:
        optimizer.zero_grad()
        outputs = model(features)
        target = target.view(-1,1)
        loss = criterion(outputs,target)
        loss.backward()
        optimizer.step()
        training_loss += loss.item()
    epoch_loss = round((training_loss/len(train_loader)),4)
    losses[f'epoch{epoch}']=epoch_loss
    train_losses.append(epoch_loss)
    model.eval()
    test_loss=0.0
    for features,target in test_loader:
        outputs=model(features)
        target = target.view(-1,1)
        loss = criterion(outputs,target)
        test_loss +=loss.item()
    avg_loss = round(test_loss/len(test_loader),4)
    test_losses.append(avg_loss)
    if avg_loss<best_val_loss:
        best_val_loss = avg_loss
        trigger_times = 0
    else:
        trigger_times += 1
        if trigger_times >=patience:
            print('Early Stopping')
            break

    
```

```python
plt.plot(train_losses,label='Train_loss')
plt.plot(test_losses,label = 'Test loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train vs Test loss')
plt.legend()
plt.show()

print(min(losses.values()),min(losses,key=losses.get))
```
